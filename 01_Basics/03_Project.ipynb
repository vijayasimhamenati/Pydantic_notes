# solution

```python
from pydantic import BaseModel, ValidationError
from datetime import date

class AutoMobiles(BaseModel):
    manufacturer : str
    series_name : str
    type_ :  str
    is_electric : bool = False
    manufactured_date : date 
    base_msrp_usd : float
    vin : str
    number_of_doors : int =4
    registration_country : str | None = None
    license_plate : str | None = None
```

```python
vijay_car = AutoMobiles.model_validate(data)
vijay_car
```

```python
vijay_car.model_fields_set
```

```python
vijay_dict = vijay_car.model_dump()
vijay_dict
```

```python
vijay_dict == data_expected_serialization
```

```python
vijay_car2 = AutoMobiles.model_validate_json(data_json)
```
